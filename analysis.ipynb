```python
# Import necessary libraries
import pandas as pd
from collections import Counter

# Load the data from the CSV file
data = pd.read_csv('smart_locks.csv')

# Function to calculate the relative ranking of brands
def calculate_relative_ranking(data):
    # Calculate the total number of SKUs
    total_skus = len(data)

    # Initialize a dictionary to store the ranking of each brand
    brand_ranking = {}

    # Iterate over each brand
    for brand in data['Brand'].unique():
        # Filter the data for the current brand
        brand_data = data[data['Brand'] == brand]

        # Calculate the total number of SKUs for the current brand
        brand_skus = len(brand_data)

        # Calculate the average rank for the current brand
        brand_rank = (brand_skus / total_skus) * 100

        # Store the ranking in the dictionary
        brand_ranking[brand] = brand_rank

    return brand_ranking

# Calculate the relative ranking of brands
brand_ranking = calculate_relative_ranking(data)

# Function to calculate the relative rating of brands
def calculate_relative_rating(data):
    # Initialize a dictionary to store the rating of each brand
    brand_rating = {}

    # Iterate over each brand
    for brand in data['Brand'].unique():
        # Filter the data for the current brand
        brand_data = data[data['Brand'] == brand]

        # Calculate the average rating for the current brand
        brand_rating[brand] = brand_data['Rating'].mean()

    return brand_rating

# Calculate the relative rating of brands
brand_rating = calculate_relative_rating(data)

# Function to calculate the price distribution of SKUs
def calculate_price_distribution(data):
    # Initialize a dictionary to store the price distribution
    price_distribution = {
        '<INR 4999': 0,
        'INR 3000-4999': 0,
        'INR 5000-9999': 0,
        'INR 10000-14999': 0,
        'INR 15000-19999': 0,
        'Greater than 20000': 0
    }

    # Iterate over each SKU
    for sku in data['Price']:
        if sku < 4999:
            price_distribution['<INR 4999'] += 1
        elif sku >= 3000 and sku <= 4999:
            price_distribution['INR 3000-4999'] += 1
        elif sku >= 5000 and sku <= 9999:
            price_distribution['INR 5000-9999'] += 1
        elif sku >= 10000 and sku <= 14999:
            price_distribution['INR 10000-14999'] += 1
        elif sku >= 15000 and sku <= 19999:
            price_distribution['INR 15000-19999'] += 1
        else:
            price_distribution['Greater than 20000'] += 1

    return price_distribution

# Calculate the price distribution of SKUs
price_distribution = calculate_price_distribution(data)

# Print the results
print("Number of Brands:", len(data['Brand'].unique()))
print("Count of SKUs per Brand:", Counter(data['Brand']))
print("Relative Ranking of Brands:", brand_ranking)
print("Relative Rating of Brands:", brand_rating)
print("Price Distribution of SKUs:", price_distribution)
```
